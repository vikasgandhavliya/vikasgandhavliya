```python
x = shiviv.drop(['District', 'Total_District_Wise'], axis = 1)
y = shiviv['Total_District_Wise'] 
```

```python
 x.shape 
```

```python
y.shape
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
model = LogisticRegression() 
model.fit(X_train, y_train) 
```

```python
 y_pred = model.predict(X_test) 
```

```python
print("Training Accuracy :", model.score(X_train, y_train)) 
print("Testing Accuracy :", model.score(X_test, y_test)) 
```

```python
from sklearn import metrics 
from sklearn.metrics import accuracy_score 
```

```python
print('''\n*"Classification report for classifier %s:\n%s\n''' % (model,metrics.classification_report(y_test,y_pred)))
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
model1 = DecisionTreeRegressor(max_depth=6)
model1.fit(x_train, y_train)
```

```python
y_pred = model1.predict(x_test)
```

```python
print("Training Accuracy :", model1.score(x_train,y_train))
print("Testing Accuracy :", model1.score(x_test, y_test))
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
classifier = GaussianNB()
classifier.fit(x_train, y_train)
```

```python
y_pred =classifier.predict(x_test)
```

```python
print("Training Accuracy :", classifier.score(x_train, y_train))
print("Testing Accuracy :",classifier.score(x_test,y_test))
```

```python
print('''\n*"Classification report for classifier %s:\n%s\n''' % (model,metrics.classification_report(y_test,y_pred)))
```
